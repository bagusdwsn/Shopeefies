```python
data = pd.read_csv("data/raw_scraped.csv", sep=',', encoding='latin1')
data['review'] = data['review'].astype(str)
data.head(10)
```

```python
# Create a new column for sentiment labels
data['sentimen'] = ''
# Iterate through the dataset and assign sentiment labels
for index, row in data.iterrows():
    rating = row['rating']
    
    if rating >3:
        data.at[index, 'sentimen'] = 'positif'
    else:
        data.at[index, 'sentimen'] = 'negatif'
data.head()
```

```python
# proses case folding 
def casefolding(review):
    if isinstance(review, str):
        review = review.lower()
    return review

data['review'] = data['review'].apply(casefolding)
data.head(10)
```

```python
import re
import string
import emoji

def cleansing(review):
    if isinstance(review, str):
        review = review.strip(" ")
        review = re.sub('[^a-zA-Z]', ' ', review)
        review = re.sub(r'[?|$|.|!_:")(-+,]', '', review)
        review = re.sub(r'\d+', '', review)
        review = re.sub(r"\b[a-zA-Z]\b", "", review)
        review = re.sub('\s+', ' ', review)
        review = emoji.demojize(review)  # Convert emojis to text representation
    return review

data['review'] = data['review'].apply(cleansing)
data.head(10)
```

```python
from nltk.tokenize import word_tokenize
#NLTK word tokenize
def word_tokenize_wrapper(text):
    if isinstance(text, str):
        return nltk.tokenize/word_tokenize(text)
    else:
        return text
data['review'] = data['review'].apply(word_tokenize_wrapper)
data.head()
```

```python
normalizad_word = pd.read_csv("normalisasi.csv")

normalizad_word_dict = {}

for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 

def normalized_term(document):
    if isinstance(document, str):
        return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]
    else:
        return document

data['review'] = data['review'].apply(normalized_term)

data['review'].head(20)
```

```python
import string
from sklearn.pipeline import Pipeline
import pandas as pd
import numpy as np
import re
```

```python
data_clean = pd.read_csv('shopeebersih.csv', encoding='latin1')
data_clean.head()
```

```python
# Example dataset of text and sentiment labels
data_clean.dropna(inplace=True)
texts = data_clean['review']
labels = data_clean['sentimen']
data_clean
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Example dataset of text and sentiment labels
texts = data_clean['review']
labels = data_clean['sentimen']

# Preprocess and tokenize the text data
# Apply any necessary text cleaning or normalization techniques

# Vectorize the text data
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Initialize KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Train the KNN model
knn.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = knn.predict(X_test)

# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import pickle

# Save the trained model to a file
filename = 'predict_sentimentv2.pkl'
with open(filename, 'wb') as file:
    pickle.dump((knn, vectorizer), file)
```
